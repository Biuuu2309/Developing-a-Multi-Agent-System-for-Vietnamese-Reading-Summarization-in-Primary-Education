```python
import pandas as pd
import re

txt_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 2\Ket noi tri thuc\doan van.txt"
xlsx_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 2\Ket noi tri thuc\outputdv.xlsx"

# Đọc file txt
with open(txt_path, "r", encoding="utf-8") as f:
    text = f.read()

# Tách đoạn:
# \n\s*\n  -> dòng trống có thể chứa khoảng trắng
paragraphs = [p.strip() for p in re.split(r"\n\s*\n", text) if p.strip()]

# Mỗi đoạn là một hàng
df = pd.DataFrame({
    "paragraph": paragraphs,
    "publisher": "KẾT NỐI TRI THỨC"
})

# Ghi ra file Excel
df.to_excel(xlsx_path, index=False)

print(f"Đã chuyển {len(df)} đoạn văn sang {xlsx_path}")
```

```python
import pandas as pd
import re

txt_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 5\Chan troi sang tao\van ban.txt"
xlsx_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 5\Chan troi sang tao\outputvb.xlsx"

# Đọc file txt
with open(txt_path, "r", encoding="utf-8") as f:
    text = f.read()

# Tách các đoạn văn (dòng trống có thể chứa khoảng trắng)
paragraphs = [p.strip() for p in re.split(r"\n\s*\n", text) if p.strip()]

titles = []
contents = []

for para in paragraphs:
    lines = [line.strip() for line in para.splitlines() if line.strip()]
    
    # Dòng đầu là tiêu đề
    title = lines[0]
    
    # Các dòng sau là nội dung
    content = "\n".join(lines[1:]) if len(lines) > 1 else ""

    titles.append(title)
    contents.append(content)

# Tạo DataFrame
df = pd.DataFrame({
    "title": titles,
    "content": contents,
    "publisher": "CHÂN TRỜI SÁNG TẠO"
})

# Ghi ra Excel
df.to_excel(xlsx_path, index=False)

print(f"Đã chuyển {len(df)} đoạn văn sang file {xlsx_path}")
```
