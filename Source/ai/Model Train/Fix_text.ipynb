```python
import pandas as pd
import re

txt_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 5\Chan troi sang tao\van ban.txt"
xlsx_path = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset grade 5\Chan troi sang tao\outputvb.xlsx"

# Đọc file txt
with open(txt_path, "r", encoding="utf-8") as f:
    text = f.read()

# Tách các đoạn văn (dòng trống có thể chứa khoảng trắng)
paragraphs = [p.strip() for p in re.split(r"\n\s*\n", text) if p.strip()]

titles = []
contents = []

for para in paragraphs:
    lines = [line.strip() for line in para.splitlines() if line.strip()]
    
    # Dòng đầu là tiêu đề
    title = lines[0]
    
    # Các dòng sau là nội dung
    content = "\n".join(lines[1:]) if len(lines) > 1 else ""

    titles.append(title)
    contents.append(content)

# Tạo DataFrame
df = pd.DataFrame({
    "title": titles,
    "content": contents,
    "publisher": "CHÂN TRỜI SÁNG TẠO"
})

# Ghi ra Excel
df.to_excel(xlsx_path, index=False)

print(f"Đã chuyển {len(df)} đoạn văn sang file {xlsx_path}")
```

```python
import re

def normalize_text(text):
    text = re.sub(r"([.,!?;:()])", r" \1 ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

for i in data_test['content']:
    print(normalize_text(i))

from vncorenlp import VnCoreNLP

rdrsegmenter = VnCoreNLP("../VnCoreNLP-master/VnCoreNLP-1.2.jar", annotators="wseg")

def word_segment(text):
    return rdrsegmenter.tokenize(text)
```

```python
from collections import defaultdict
import pandas as pd

data = pd.read_excel(r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\Data TX.xlsx")
data_test = data
grade_vocab = defaultdict(set)

for _, row in data_test.iterrows():
    tokens = word_segment(normalize_text(row["content"]))
    for sent in tokens:
        for word in sent:
            grade_vocab[row["grade"]].add(word.lower())

len(grade_vocab)
```

```python
import os

output_dir = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\grade_vocab"
os.makedirs(output_dir, exist_ok=True)

from collections import defaultdict

cumulative_vocab = defaultdict(set)

sorted_grades = sorted(grade_vocab.keys())

current_vocab = set()

for grade in sorted_grades:
    current_vocab |= grade_vocab[grade]   # union
    cumulative_vocab[grade] = current_vocab.copy()

for grade, vocab in cumulative_vocab.items():
    file_path = os.path.join(output_dir, f"grade_{grade}.txt")
    with open(file_path, "w", encoding="utf-8") as f:
        for word in sorted(vocab):
            f.write(word + "\n")

```
