```python
import re

def normalize_text(text):
    text = re.sub(r"([.,!?;:()])", r" \1 ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

for i in data_test['content']:
    print(normalize_text(i))

from vncorenlp import VnCoreNLP

rdrsegmenter = VnCoreNLP("../VnCoreNLP-master/VnCoreNLP-1.2.jar", annotators="wseg")

def word_segment(text):
    return rdrsegmenter.tokenize(text)
```

```python
from collections import defaultdict
import pandas as pd

data = pd.read_excel(r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\Data TX.xlsx")
data_test = data
grade_vocab = defaultdict(set)

for _, row in data_test.iterrows():
    tokens = word_segment(normalize_text(row["content"]))
    for sent in tokens:
        for word in sent:
            grade_vocab[row["grade"]].add(word.lower())

len(grade_vocab)
```

```python
import os

output_dir = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\grade_vocab"
os.makedirs(output_dir, exist_ok=True)

from collections import defaultdict

cumulative_vocab = defaultdict(set)

sorted_grades = sorted(grade_vocab.keys())

current_vocab = set()

for grade in sorted_grades:
    current_vocab |= grade_vocab[grade]   # union
    cumulative_vocab[grade] = current_vocab.copy()

for grade, vocab in cumulative_vocab.items():
    file_path = os.path.join(output_dir, f"grade_{grade}.txt")
    with open(file_path, "w", encoding="utf-8") as f:
        for word in sorted(vocab):
            f.write(word + "\n")

```

```python
from vncorenlp import VnCoreNLP
import pandas as pd
vncorenlp = VnCoreNLP(
    r"../../VnCoreNLP-master/VnCoreNLP-1.2.jar",  # chỉnh đúng path của bạn
    annotators="wseg",
    max_heap_size='-Xmx2g'
)

df = pd.read_excel(r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset abstractive\DATA_DG_main.xlsx")

def count_sentences_and_words(text):
    if not isinstance(text, str) or text.strip() == "":
        return 0, 0

    sentences = text.split(".")
    sentences = [s.strip() for s in sentences if len(s.strip()) > 0]

    word_count = 0
    for sent in sentences:
        tokens = vncorenlp.tokenize(sent)
        word_count += sum(len(s) for s in tokens)

    return len(sentences), word_count
df_g1 = df[(df["grade"] == 3) & (df["publisher"].notna())].copy()
df_g1["num_sentences"] = 0
df_g1["num_words"] = 0

for idx, row in df_g1.iterrows():
    sents, words = count_sentences_and_words(row["content"])
    df_g1.at[idx, "num_sentences"] = sents
    df_g1.at[idx, "num_words"] = words
max_sent_row = df_g1.loc[df_g1["num_sentences"].idxmax()]

print("📌 Văn bản có nhiều câu nhất (grade=1)")
print("Số câu:", max_sent_row["num_sentences"])
print("Số từ:", max_sent_row["num_words"])
print("Nội dung:\n", max_sent_row["content"])

max_word_row = df_g1.loc[df_g1["num_words"].idxmax()]

print("📌 Văn bản có nhiều từ nhất (grade=1)")
print("Số câu:", max_word_row["num_sentences"])
print("Số từ:", max_word_row["num_words"])
print("Nội dung:\n", max_word_row["content"])
```

```python
import pandas as pd

# ===== 1. Load dữ liệu =====
INPUT_XLSX = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset abstractive\DATA_DG_main_2.xlsx"
OUTPUT_XLSX = r"E:\Project_NguyenMinhVu_2211110063\Source\datasets\dataset abstractive\DATA_DG_main_3.xlsx"

df = pd.read_excel(INPUT_XLSX)

# ===== 2. Lọc grade = 1 & publisher null =====
mask_g1 = (df["grade"] == 3) & (df["publisher"].isna())
df_g1 = df[mask_g1].copy()

# ===== 3. Hàm đếm câu & từ (NHANH – ỔN ĐỊNH) =====
def count_fast(text):
    if not isinstance(text, str) or not text.strip():
        return 0, 0

    # đếm câu (đủ chính xác cho giáo dục)
    sentences = text.replace("!", ".").replace("?", ".").split(".")
    num_sentences = len([s for s in sentences if s.strip()])

    # đếm từ
    num_words = len(text.split())

    return num_sentences, num_words

# ===== 4. Xác định các dòng cần xóa =====
drop_idx = []

for idx, row in df_g1.iterrows():
    s, w = count_fast(row["content"])
    if s > 45 or w > 650:
        drop_idx.append(idx)

# ===== 5. Xóa dòng trong df gốc =====
df = df.drop(index=drop_idx).reset_index(drop=True)

# ===== 6. Lưu file =====
df.to_excel(OUTPUT_XLSX, index=False)

print("✅ Hoàn tất lọc dữ liệu")
print(f"🗑️ Số dòng bị xóa: {len(drop_idx)}")
```
